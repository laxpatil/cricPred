```python
import graphlab as gl
```

```python
data= gl.SFrame("../Instantaneous features/BatsmanFeaturesUpdated.csv")
```

```python
data
```

```python
kmeans_model = gl.kmeans.create(data,features=['Home Run Hitting Ability', 'Milestone Reaching Ability','Batting Average','Strike Rate'], num_clusters=5)
```

```python
kmeans_model.summary()
```

```python
kmeans_model['cluster_info'].print_rows(num_columns=5, max_row_width=1000,
                                        max_column_width=1000)
```

```python
kresult=kmeans_model['cluster_id']
```
