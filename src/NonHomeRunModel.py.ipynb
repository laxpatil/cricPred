```python
import graphlab as gl
```

```python
batsman= gl.SFrame("../Instantaneous features/batsmanClusters.csv")
```

```python
batsman
```

```python
historical=gl.SFrame("../Historical features/HistoricalFeatures.csv")
```

```python
segments=[]
for i in range(10):
    seg=gl.SFrame("../Segment Features/Segment"+str(i+1)+".csv")
    segments.append(seg)
```

```python
segments[0]
```

### Get the combined features

```python
all_features=[]
for i in range(10):
    combined=historical.join(segments[i], on='Team')
    combined= combined.join(batsman, on='Batsman')
    all_features.append(combined)
```

```python
modelFeatures=[]
for feature in all_features:
    mdFeature=feature.select_columns(['Runs Scored','Wickets Lost','Got All Out', 'Runs Conceded','Opponent Wickets Taken','Opponent All Out','Total Runs Scored','Balls Faced', 'R0','W0','R1','W1','Target','ClusterID','Non Home Runs'])
    modelFeatures.append(mdFeature)
```

#### Select features columns

```python
modelFeatures[0]
```

```python
train_data=[]
test_data=[]
for i in range(10):
    train, test=modelFeatures[i].random_split(0.8,seed=0)
    train_data.append(train)
    test_data.append(test)
```

```python
models=[]
for i in range(10):
    linearModel= gl.linear_regression.create(train_data[i], target='Non Home Runs')
    models.append(linearModel)
```

```python
for i in range(10):
    models[i].summary()
```

```python
models[6].evaluate(test_data[6])
```

```python
for i in range(10):
    print(models[i].evaluate(test_data[i]))
```
